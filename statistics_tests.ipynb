```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("data-69d3709338fdb705334215.csv")
df
```

```python
df_mnts = df[[
    'MntBakedProds',
    'MntFishProducts',
    'MntFruits',
    'MntMeatProducts', 
    'MntFishProducts',
    'MntSweetProducts',
    'MntBakedProds'
]
]

target = df_mnts.mean()
p_value = stats.ttest_1samp(df_mnts, target)

alpha = 0.05
if p_value < alpha:
    print("on ne rejet H0")
else:
    print("on ne rejet pes H0")
```

```python
from scipy.stats import chi2_contingency

accepted_campaign = df[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']]
obs = np.array([accepted_campaign, df['Country']])
res = chi2_contingency(obs)
res.statistic
```
