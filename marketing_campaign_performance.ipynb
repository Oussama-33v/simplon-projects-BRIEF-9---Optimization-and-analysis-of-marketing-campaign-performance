```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as act
```

```python
#couvert the data into a data frame
df = pd.read_csv("data-69d3709338fdb705334215.csv")
df
```

```python
#analyse the variable "Education"
df['Education'].value_counts()
```

```python
# a bar chart for this variable
plt.figure(figsize=(11,5))
df['Education'].value_counts().plot.bar()
plt.title("deff education count", c = 'k', fontsize = 16)
```

```python
#analyse the variable income
df[' Income '].describe()
```

```python
df[' Income '].sum()
```

```python
plt.figure(figsize=(11,5))
sns.histplot(data = df, x = df[' Income '])
plt.title("description of income values", c = 'k', fontsize = 26)
plt.ylabel("")
```

```python
plt.figure(figsize=(11,5))
sns.boxplot(data = df, x = df[' Income '])
```

```python
#analyse the variable country
df['Country'].value_counts()
```

```python
plt.figure(figsize=(11,5))
df['Country'].value_counts().plot.bar()
plt.title("count of customers by country", c = 'k', fontsize = 16)
```

Bar charts for the number of customers who accepted each campaign.

```python
# Melt the data to long format

df_melted = df.melt(value_vars=['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5'], 
                    var_name='Campaign', value_name='Accepted')

plt.figure(figsize=(12, 6))
sns.countplot(data=df_melted, x='Campaign', hue='Accepted')
plt.title('Campaign Acceptance Comparison (0 vs 1)')
plt.ylabel("")
```

```python
plt.figure(figsize=(14, 5))
matrix = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].corr()

sns.heatmap(matrix, cmap = 'coolwarm', fmt = '.2f', annot = True)
```

TOTAL SALES FOR EVERY COUNTRY

```python
plt.figure(figsize=(11,5))
sns.barplot(data = df, x = df['Country'], y = df[' Income '], errorbar= None, color ='#EFBF04')
plt.title("yearly income by country", c = 'k', fontsize = 20)
plt.xlabel("Countrys")
```
